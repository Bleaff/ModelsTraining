```python
train_dict = []
with open('/media/bleaf/4f298626-e60e-4eac-b4b8-3a01fc666bf3/bleaf/Datasets/crowd_human_dataset/annotation_train.odgt') as anno_train:
    for line in anno_train:
        train_dict.append(yaml.safe_load(line))
```

```python
path_for_labels = '/media/bleaf/4f298626-e60e-4eac-b4b8-3a01fc666bf3/bleaf/Datasets/crowd_human_dataset/train/labels/'
path_for_images ='/media/bleaf/4f298626-e60e-4eac-b4b8-3a01fc666bf3/bleaf/Datasets/crowd_human_dataset/train/images/'
```

```python
path_for_labels = '/media/bleaf/4f298626-e60e-4eac-b4b8-3a01fc666bf3/bleaf/Datasets/crowd_human_dataset/valid/labels/'
path_for_images ='/media/bleaf/4f298626-e60e-4eac-b4b8-3a01fc666bf3/bleaf/Datasets/crowd_human_dataset/valid/images/'
os.makedirs(path_for_labels, exist_ok=True)
for i, img_l in enumerate(train_dict):
    img = cv2.imread(os.path.join(path_for_images, f'{img_l["ID"]}.jpg'))
    h, w, c= img.shape
    with open(os.path.join(path_for_labels, f"{img_l['ID']}.txt"), 'w') as lt:
        for box in img_l['gtboxes']:
            x, y, wi, he = box['vbox']
            x = x / w + wi / (2 * w)
            y = y / h + he / (2 * h)
            wi = wi / w
            he = he / h
            lt.write(f'0 {x} {y} {wi} {he}\n')
    print(f'Done {i}/{len(train_dict)}')
```

```python
def  read_file(file_name):
    with  open(file_name,  'r')  as f:
        read  =  f.read()
    splitted  = read.split('\n')
    splitted  = [i.split()  for  i  in  splitted]
    # print(splitted)
    splitted  = [i  for i  in  splitted if len(i)]
    splitted  = [[int(c),  float(x), float(y),  float(h),  float(w)]  for  c,  x, y,  h, w  in  splitted]
    return splitted

def get_points(xc, yc, h, w):
    x1, y1 = xc - w/2, yc -h/2
    x2, y2 = xc + w/2, yc +h/2
    return [x1, y1, x2, y2]

def show_boxes(id, path):
    labels = read_file(os.path.join(path, 'labels', f'{id}.txt'))
    img = cv2.imread(os.path.join(path, 'images', f'{id}.jpg'))
    h, w, _ = img.shape
    for label in labels:
        c, x, y, hl, wl  = label
        x1, y1, x2, y2  =  get_points(x, y,  wl, hl)
        img = cv2.rectangle(img, (int(x1*w), int(y1*h)), (int(x2*w), int(y2*h)), (255, 0, 0))
    plt.imshow(img)
```
