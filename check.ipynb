```python
ipath = '/home/bleaf/Downloads/people/images'
lpath  = '/home/bleaf/Downloads/people/labels/all'
```

```python
lid = [{'image':img, 'name':img.split('.')[0], 'ext':img.split('.')[1]} for img in os.listdir(ipath)]
li =  os.listdir(ipath)
lld = [{'label':lbl, 'name':lbl.split('.')[0]} for lbl in os.listdir(lpath)]
ll = os.listdir(lpath)
```

```python
names= [img.split('.')[0] for img in li]
lnames = [img.split('.')[0] for img in ll]
exts = [img.split('.')[1] for img in li]
```

```python
names  = set(names)
lnames  =  set(lnames)
```

```python
print(len(names))
```

```python
print(len(lnames))
```

```python
intrsect = names.intersection(lnames)
```

```python
save_img  = 'dataset_from_internet/images'
save_lbl  = 'dataset_from_internet/labels'

os.makedirs(save_img, exist_ok=True)
os.makedirs(save_lbl, exist_ok=True)
```

```python
images_to_add = []
labels_to_add = []


for name in intrsect:
    for img in lid:
        if  name == img['name']:
            images_to_add.append(img['image'])
    for lbl in lld:
        if  name == lbl['name']:
            labels_to_add.append(lbl['label'])


```

```python
labels_to_add
```

```python
for img,  label in  zip(images_to_add,  labels_to_add):
    shutil.copy(os.path.join(ipath, img), os.path.join(save_img, img))
    shutil.copy(os.path.join(lpath, label), os.path.join(save_lbl, label))
```

```python
import os

def  read_file(file_name):
    with  open(file_name,  'r')  as f:
        read  =  f.read()
    splitted  = read.split('\n')
    splitted  = [i.split()  for  i  in  splitted]
    # print(splitted)
    splitted  = [i  for i  in  splitted if len(i)]
    splitted  = [[int(c),  float(x), float(y),  float(h),  float(w)]  for  c,  x, y,  h, w  in  splitted]
    return splitted
read_file('/home/bleaf/train_people/test.txt')
```

```python
path =  '/home/bleaf/train_people/dataset/train/labels'
lst_path  = os.listdir(path)
to_delete   = []
for txt  in lst_path:
    readen =  read_file(os.path.join(path,  txt))
    for line in readen:
        if  line[0] != 0:
            to_delete.append(txt)
```

```python
ipath  = '/home/bleaf/train_people/dataset/train/images/'
not_found =  []
for  i in to_delete:
    try:
        name, ext   =  i.split('.')
        os.remove(os.path.join(ipath,  f'{name}.jpg'))
        os.remove(os.path.join(path, i))
        print(i,  'deleted')
    except Exception  as e:
        print(e)
        not_found.append(i)
```
